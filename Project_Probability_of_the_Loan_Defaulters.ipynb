```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:\Data Science ML AI\Python files\Projects\Project_Probability of the Loan Defaulters\Loan_defaulters_dataset.csv")
df.head(10)
```

```python
# probability of  fico score greater than 700
p_a = df[df['fico'].astype(float) > 700].shape[0]/df.shape[0]
print("probability of fico score greater than 700 :", round(p_a,2))
```

```python
# probability of purpose == debt_consolidation
p_b = df[df['purpose'] == 'debt_consolidation'].shape[0]/df.shape[0]
print("probability of purpose == debt_consolidation :", round(p_b,2))
```

```python
# Create new dataframe for condition ['purpose']== 'debt_consolidation' 
df1 = df[df['purpose'] == 'debt_consolidation']
df1.head(10)
```

```python
# Calculate the P(A|B)
#for the event purpose == 'debt_consolidation' given 'fico' credit score is greater than 700
p_a_b = df[df['fico'].astype(float) > 700].shape[0]/df1.shape[0]
print("P(A|B) is:", str(round(p_a_b,2)))
```

```python
# Check whether the P(A) and P(B) are independent from each other
result = (p_a == p_a_b)
print(result)
```

```python
# probability of paid_back_loan is Yes
prob_pd = df[df['paid.back.loan'] == 'Yes'].shape[0]/df.shape[0]
print("probability of paid back loan is:", round(prob_pd,2))
```

```python
# probability of credit.policy is Yes
prob_cs = df[df['credit.policy'] == 'Yes'].shape[0]/df.shape[0]
print("probability of credit policy is:", round(prob_cs,2))
```

```python
# create new dataframe for paid.back.loan == 'Yes'
new_df = df[df['paid.back.loan'] == 'Yes']
new_df.head(10)
```

```python
#calculate the probablityp(B|A) for the event paid.back.loan == 'Yes' given credit.policy == 'Yes'
prob_pd_cs = new_df[new_df['credit.policy'] == 'Yes'].shape[0]/new_df.shape[0]
print(round(prob_pd_cs,2))
```

```python
# bayes theorem to calculate P(A|B)
bayes = (prob_pd_cs*prob_pd)/prob_cs
print("Output of Byes theorem is:", round(bayes,4))
```
